<h1>Literacy rate of the World and Pakistan with time</h1>

We have obtained a data sheet from the web, through World Bank, describing the literacy rate in youth of various countries of the world. We will draw a line plot of Pakistan and The world through the years.

<h3>Pakistan</h3>

We will draw a line plot out of the data obtained from the World Bank, to have an insight on how th literacy rate has evolved in Pakistan.

First we import the pandas library, which we will use to read the excel file we have obtained from the web.

```python
import pandas as pd
import numpy as np
```

We will then read the excel file into a dataframe.

```python
df = pd.read_excel("Data Task 1.xlsx")
```

Now we will read the data from the <b>"Pakistan"</b> row from the dataframe.

```python
Pakistan = "Pakistan"
Pak_row = df[df.iloc[:, 0] == Pakistan]
```

To plot a graph, we will drop first two non-numeric rows, which contain the name and abbreviation of the country.

```python
row_data = Pak_row.iloc[0].drop(labels=[df.columns[0],df.columns[1]])
```

We will now convert the data two numeric form instead of string.

```python
row_data = pd.to_numeric(row_data)
row_data = row_data.dropna()
```

Now we will import Pyplot from Matplotlib to plot the line plot from the data we have obtained.

```python
from matplotlib import pyplot as plt
```

Now the final task, plotting the graph.

```python
plt.plot(row_data.index, row_data.values, marker='o', linestyle='-', color='b')
plt.xticks(rotation=45, ha='right')
plt.title('Line plot for Pakistan')
plt.xlabel('Values')
plt.ylabel('Frequency')
y_ticks = np.linspace(0.5,1, 6)  
plt.yticks(y_ticks)
plt.show()
```

<h3>The World</h3>

Now we will repeat the same steps but for the data from the world.

```python
World = "World"
World_row = df[df.iloc[:, 0] == World]
row_data2 = World_row.iloc[0].drop(labels=[df.columns[0],df.columns[1]])
row_data2 = pd.to_numeric(row_data2)
row_dataw = row_data2
row_data2 = row_data2[::2]
```

```python
plt.plot(row_data2.index, row_data2.values, marker='o', linestyle='-', color='b')
plt.xticks(rotation=45, ha='right')
plt.title('Line plot for World')
plt.xlabel('Values')
plt.ylabel('Frequency') 
y_ticks = np.linspace(0.5,1, 6)  # Defines where ticks will appear
plt.yticks(y_ticks)
plt.show()
```

<h2>Comparison</h2>
Now we will overlay the two graphs over each other to better see the difference between Pakistan and The world

```python
common_index = row_dataw.index.intersection(row_data.index)
common_index = common_index[20:]

aligned_row_data2 = row_dataw.loc[common_index]
aligned_row_data = row_data.loc[common_index]

aligned_row_data = aligned_row_data.dropna()

plt.plot(aligned_row_data2.index, aligned_row_data2.values, marker='o', linestyle='-', color='b', label="World")
plt.plot(aligned_row_data.index, aligned_row_data.values, marker='o', linestyle='-', color='g', label='Pakistan')

plt.xticks(rotation=45, ha='right')
plt.title('Line plot for World and Pakistan')
plt.xlabel('Values')
plt.ylabel('Frequency') 

y_ticks = np.linspace(0.5, 1, 6)
plt.yticks(y_ticks)

plt.legend()
plt.show()
```
